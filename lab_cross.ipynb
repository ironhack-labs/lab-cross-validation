```python
data['SeniorCitizen'] = data['SeniorCitizen'].astype(object)
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
```

```python
#checking the null values

nulls_data = data.isna().sum()*100 / len(data)
nulls_data
```

```python
mean = data['TotalCharges'].mean()
data['TotalCharges'] = data['TotalCharges'].fillna(mean)
```

```python
#Spliting the data into numerical and categorical

numerical = data.select_dtypes(exclude='object')
categorical = data.select_dtypes(include='object')
y = categorical['Churn']
categorical = categorical.drop(['Churn'], axis=1)
```

```python
#scaling the numerical data

def standard_scaler(data):
    columns = data.columns
    data_copy = pd.DataFrame(StandardScaler().fit(data).transform(data))
    data_copy.columns = columns
    return data_copy

numerical_scaled = standard_scaler(numerical)
numerical_scaled
```

```python
# Encoding the categorical data

columns_to_encode = list(categorical.columns)
categorical_encoded = pd.get_dummies(categorical[columns_to_encode])
```
