```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import seaborn as sns
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import TomekLinks
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('Customer-Churn.csv')
```

```python
df
```

```python
# Check data types and transform if necessary 

df.dtypes
```

```python
df['SeniorCitizen'] = df['SeniorCitizen'].astype(object)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
# 4. Check for null values in the dataframe. Replace the null values.
nulls_df = df.isna().sum()*100 / len(df)
nulls_df
```

```python
mean = df['TotalCharges'].mean()
df['TotalCharges'] = df['TotalCharges'].fillna(mean)
```

```python
nulls_df = df.isna().sum()*100 / len(df)
nulls_df
```

```python
# 5. Split the data into numerical and catagorical. 
numerical = df.select_dtypes(exclude='object')
categorical = df.select_dtypes(include='object')
y = categorical['Churn']
categorical = categorical.drop(['Churn'], axis=1)
```

```python
numerical.dtypes
```

```python
categorical.dtypes
```

```python
# Scale the numeric columns using the standard scaler
def standard_scaler(df):
    columns = df.columns
    df_copy = pd.DataFrame(StandardScaler().fit(df).transform(df))
    df_copy.columns = columns
    return df_copy

numerical_scaled = standard_scaler(numerical)
numerical_scaled
```

```python
# Encode the categorical features using One-Hot Encoding.
columns_to_encode = list(categorical.columns)
categorical_encoded = pd.get_dummies(categorical[columns_to_encode])
```

```python
categorical_encoded
```

```python
# Concatenate numerical and categorical back together again for your X dataframe. Designate the Target as y.

X = pd.concat([numerical_scaled, categorical_encoded], axis=1)
```

```python
X
```

```python
# Apply SMOTE for upsampling the data

sns.countplot(x = y, data = X, palette = 'Set3')
```

```python
sm = SMOTE()
X_smote, y_smote = sm.fit_resample(X, y)
sns.countplot(x = y_smote, data = X_smote, palette = 'Set3')

```

```python
# Split the data into a training set and a test set.
X_smote_train, X_smote_test, y_smote_train, y_smote_test = train_test_split(X_smote,y_smote, test_size=0.4, random_state=100)
```

```python
# Use logistic regression to fit the model and compute the accuracy of the model.

model_smote = LogisticRegression(max_iter=1000).fit(X_smote_train, y_smote_train)
y_smote_pred = model_smote.predict(X_smote_test)
print('The Logistic Regression accuracy (SMOTE) is: ', model_smote.score(X_smote_test, y_smote_test))
```

```python
# Use decision tree classifier to fit the model and compute the accuracy of the model.
# Compare the accuracies of the two models.

tree_model_smote = DecisionTreeClassifier()
tree_model_smote.fit(X_smote_train, y_smote_train)
print('The Decision Tree accuracy (SMOTE) is: ', tree_model_smote.score(X_smote_test,y_smote_test))
```

```python
# Apply TomekLinks for downsampling
tl = TomekLinks()
X_tl, y_tl = tl.fit_resample(X, y)
sns.countplot(x = y_tl, data = X_tl, palette = 'Set3')
```

```python
# Split the data into a training set and a test set.
X_tl_train, X_tl_test, y_tl_train, y_tl_test = train_test_split(X_tl,y_tl, test_size=0.4, random_state=100)
```

```python
# Use logistic regression to fit the model and compute the accuracy of the model.

model_tl = LogisticRegression(max_iter=1000).fit(X_tl_train, y_tl_train)
y_tl_pred = model_tl.predict(X_tl_test)
print('The Logistic Regression accuracy (TL) is: ', model_tl.score(X_tl_test, y_tl_test))
```

```python
# Use decision tree classifier to fit the model and compute the accuracy of the model.
# Compare the accuracies of the two models.

tree_model_tl = DecisionTreeClassifier()
tree_model_tl.fit(X_tl_train, y_tl_train)
print('The Decision Tree accuracy (TL) is: ', tree_model_tl.score(X_tl_test,y_tl_test))
```

```python
print('The Logistic Regression accuracy (SMOTE) is: ', model_smote.score(X_smote_test, y_smote_test))
print('The Logistic Regression accuracy (TL) is: ', model_tl.score(X_tl_test, y_tl_test))
print('The Decision Tree accuracy (SMOTE) is: ', tree_model_smote.score(X_smote_test,y_smote_test))
print('The Decision Tree accuracy (TL) is: ', tree_model_tl.score(X_tl_test,y_tl_test))
```
